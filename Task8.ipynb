```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Campaign": ["Email", "Social Media", "Google Ads", "YouTube Ads", "Affiliate"],
    "Impressions": [50000, 80000, 65000, 70000, 40000],
    "Clicks": [2500, 4200, 3600, 3900, 1800],
    "Conversions": [250, 420, 340, 380, 150],
    "Cost": [10000, 15000, 18000, 20000, 8000]
}

df = pd.DataFrame(data)


df["CTR (%)"] = (df["Clicks"] / df["Impressions"]) * 100
df["Conversion Rate (%)"] = (df["Conversions"] / df["Clicks"]) * 100

print("========== MARKETING ANALYTICS REPORT ==========\n")
print(df)

print("\n========== KPI REPORT ==========")
print("Total Impressions:", df["Impressions"].sum())
print("Total Clicks:", df["Clicks"].sum())
print("Total Conversions:", df["Conversions"].sum())
print("Total Campaign Cost:", df["Cost"].sum())

best_campaign = df.loc[df["Conversions"].idxmax()]
print("\nBest Performing Campaign")
print(best_campaign)

print("\n========== SUMMARY ==========")
print(df.describe())


plt.figure(figsize=(8,5))
plt.bar(df["Campaign"], df["Clicks"])
plt.title("Campaign Click Analysis")
plt.xlabel("Campaign")
plt.ylabel("Clicks")
plt.show()


plt.figure(figsize=(8,5))
plt.plot(df["Campaign"], df["Conversions"], marker='o')
plt.title("Campaign Conversion Analysis")
plt.xlabel("Campaign")
plt.ylabel("Conversions")
plt.grid(True)
plt.show()

print("\nMarketing Analytics Completed Successfully!")
```
